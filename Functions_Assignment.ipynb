```python
#6) Explain the concept of generators in Python and how they are defined.

'''
Generators in Python are a simple and powerful way of creating iterators. They are functions that can be paused and resumed, generating values on the fly instead of creating a full list in memory at once. This makes them highly memory efficient.
Generators don’t use a return statement to give back a single value and exit. Instead, it uses the yield keyword. When yield is called, the generated value is returned to the caller. The next time the generator is called, it resumes execution from exactly where it left off.
'''
```

```python
#7) What are the advantages of using generators over regular functions?

'''
Generators are superior to regular functions for large/infinite data because they use lazy evaluation, yielding values one by one instead of storing everything in memory, leading to huge memory savings (memory efficiency). They can represent infinite data streams (like all prime numbers or Fibonacci sequence) because they don't need to store all values. Generators automatically implement iterator protocols (__iter__, __next__) with just the yield keyword, making iterator creation much cleaner than regular functions. 
'''
```

```python
# 10) What is the difference between `map ()`, `reduce()`, and `filter()` functions in Python?

'''
Function 	Purpose	Input	Output
map()	Transform	Applies a function to each element of an iterable.	A new iterable with the transformed elements (same length as the original).
filter()	Filter	Applies a function to each element and keeps only those for which the function returns True.	A new iterable with a subset of the original elements (potentially shorter length).
reduce()	Aggregate	Applies a function cumulatively to the elements of an iterable to reduce it to a single value.	A single value (e.g., sum, product, single combined string).
'''
```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(nums)
print(result)
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_list(numbers):
    return [num ** 2 for num in numbers]

nums = [1, 2, 3, 4]
print(square_list(nums))
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

fib = Fibonacci(10)

for num in fib:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for value in powers_of_two(5):
    print(value, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

with open("myfile.txt", "w") as f:
    f.write("Python\nRobot Framework\nProbability\nPattern Matching")

def read_file_line_by_line(file_path):
    with open(file_path, "r") as file:
        for line in file:
            yield line.rstrip("\n")

for line in read_file_line_by_line("myfile.txt"):
    print(line)
```

```python
#9. Write a Python program that uses `map ()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

# Convert to Fahrenheit using map()
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
#10. Create a Python program that uses `filter () ` to remove all the vowels from a given string.

def is_not_vowel(ch):
    vowels = "aeiouAEIOU"
    return ch not in vowels

def remove_vowels(text):
    return ''.join(filter(is_not_vowel, text))

# Example usage
string = "Hello World"
result = remove_vowels(string)

print("Original String:", string)
print("After removing vowels:", result)
```
